```python
#Importing the libraries:
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Keyword arguement notation:
#plt.plot(x,y, label = '2x', color = 'blue', linewidth = 2, marker = '.', linestyle = '--', markersize = 10, markeredgecolor = 'red')

#Shorthand notation of the above line:
plt.plot(x,y,'b.--', label='2x')

#Plotting the line number 2:
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'r', label = 'x^2')
plt.title("First Graph!", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 24})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize the graph:
plt.figure(figsize=(5,3), dpi = 100)

#Keyword arguement notation:
#plt.plot(x,y, label = '2x', color = 'blue', linewidth = 2, marker = '.', linestyle = '--', markersize = 10, markeredgecolor = 'red')

#Shorthand notation of the above line:
plt.plot(x,y,'b.--', label='2x')

#Plotting the line number 2:
#Selecting the interval for which we want to plot the points at:
x2 = np.arange(0,4.5,0.5)

#Plot the part of the graph as the line:
plt.plot(x2[:6],x2[:6]**2,'r', label = 'x^2')

#PLot the remainder of the graph:
plt.plot(x2[5:], x2[5:]**2, 'r--')

#Giving the title of the graph:
plt.title("First Graph!", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 24})

#Labelling the axis of the graph:
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

#Assiging the ticks of the aixs of the graph (Scale of the graph):
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#Displaying the legend of the graph:
plt.legend()

#Saving the graph:
plt.savefig('mygraph.png', dpi = 300)

#Displaying the graph:
plt.show()
```

### Bar Charts

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar(labels, values)

#Defining the patterns that we want to use to fill in the bars:
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())
#bars[0].set_hatch('*')
#bars[1].set_hatch('/')
#bars[2].set_hatch('O')

#Resizing the graph:
plt.figure(figsize = (6,4))
plt.show()
```
